```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

```

```python
df = pd.read_csv("dataset.csv")
```

```python
#2. Fix NULL values
num_cols = ["AGE","BP","CHOLESTROL","bmi","stress_level","sleep_hours"]

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Categorical columns → fill with mode
cat_cols = ["gender","family_history","smoking_status",
            "alcohol_consumption","physical_activity",
            "diabetes","hypertension"]

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#3. Fix Outliers (IQR method)

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])
```

```python
# 4. Round AGE and sleep_hours

df["AGE"] = df["AGE"].fillna(df["AGE"].median()).round().astype(int)
df["sleep_hours"] = df["sleep_hours"].fillna(df["sleep_hours"].median()).round().astype(int)
```

```python
# 5. Encode categorical columns
le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```
